```python
import pandas as pd
import re
from datetime import datetime
import calendar
```

```python
# Get 2015 rebalance dates
df_2015 = pd.read_html('https://www.thestreet.com/stock-market-news/11771386/market-holidays-2013.html')
```

```python
# Clean date column
df_2015[1].columns = df_2015[1].iloc[0, :]
df_2015[1].drop([0], axis=0, inplace=True)
df_2015[1]['date'] = list(map(lambda x: re.search('[A-z]+(\.?) \d\d?(\W+)\d{4}', x).group(), df_2015[1].iloc[:, 0]))
df_2015[1]['date'] = list(map(lambda x: datetime.strptime(x, '%B %d, %Y'), df_2015[1]['date']))
```

```python
# Select those in 2015
start_date = datetime.strptime('2015-1-1', '%Y-%m-%d')
end_date = datetime.strptime('2015-12-31', '%Y-%m-%d')
mask = (df_2015[1]['date'] >= start_date) & (df_2015[1]['date'] <= end_date)
df_2015[1] = df_2015[1].loc[mask]
```

```python
c = calendar.Calendar(firstweekday=calendar.SUNDAY)

year = 2015
# month = 2
months = [3, 6, 9, 12]
rebal_dates_2015 = []

for month in months:
    monthcal = c.monthdatescalendar(year,month)
    third_friday = [day for week in monthcal for day in week if \
                    day.weekday() == calendar.FRIDAY and \
                    day.month == month][2]
    rebal_dates_2015.append(third_friday)
```

```python
for rebal_date in rebal_dates_2015:
    print(rebal_date in df_2015[1]['date'])
```

```python
df = pd.DataFrame(data=rebal_dates_2015)
```
