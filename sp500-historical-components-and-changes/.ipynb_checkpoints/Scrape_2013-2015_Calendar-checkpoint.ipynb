```python
import pandas as pd
import re
from datetime import datetime
from datetime import timedelta
import calendar
import requests
```

```python
# Get 2015 holidays
df_2015 = pd.read_html('https://www.thestreet.com/stock-market-news/11771386/market-holidays-2013.html')
```

```python
# Clean date column
df_2015[1].columns = df_2015[1].iloc[0, :]
df_2015[1].drop([0], axis=0, inplace=True)
df_2015[1]['date'] = list(map(lambda x: re.search('[A-z]+(\.?) \d\d?(\W+)\d{4}', x).group(), df_2015[1].iloc[:, 0]))
df_2015[1]['date'] = list(map(lambda x: datetime.strptime(x, '%B %d, %Y'), df_2015[1]['date']))
```

```python
# Select those in 2015
start_date = datetime.strptime('2015-1-1', '%Y-%m-%d')
end_date = datetime.strptime('2015-12-31', '%Y-%m-%d')
mask = (df_2015[1]['date'] >= start_date) & (df_2015[1]['date'] <= end_date)
df_2015[1] = df_2015[1].loc[mask]
```

```python
def get_rebal_dates(years, holidays):
    '''
    Generate rebalance dates that are the 3rd Friday of ending month of every quarter
    Avoiding trading holidays (i.e. offset backwards by one trading date if falls on a holiday)
    '''

    c = calendar.Calendar(firstweekday=calendar.SUNDAY)

    months = [3, 6, 9, 12]
    rebal_dates = []
    
    for year in years:
        for month in months:
            monthcal = c.monthdatescalendar(year,month)
            third_friday = [day for week in monthcal for day in week if \
                            day.weekday() == calendar.FRIDAY and \
                            day.month == month][2]
            while (third_friday == holidays).any() or third_friday.isoweekday() == 6 or third_friday.isoweekday() == 7:
                third_friday += datetime.timedelta(days=1)
            
            rebal_dates.append(third_friday)
    return rebal_dates
```

```python
rebal_dates = []
```

```python
rebal_dates.extend(get_rebal_dates([2013], df_2013[0].loc[df_2013[0]['Status']=='Closed (passed)', 'date']))
```

```python
rebal_dates.extend(get_rebal_dates([2015], df_2015[1].loc[df_2015[1]['NASDAQ'] == 'CLOSED', 'date']))
```
